```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined
files_2022 = ['Data/JC-202201-citibike-tripdata.csv', 'Data/JC-202202-citibike-tripdata.csv', 'Data/JC-202203-citibike-tripdata.csv', 'Data/JC-202204-citibike-tripdata.csv', 'Data/JC-202205-citibike-tripdata.csv', 'Data/JC-202206-citibike-tripdata.csv', 'Data/JC-202207-citbike-tripdata.csv', 'Data/JC-202208-citibike-tripdata.csv', 'Data/JC-202209-citibike-tripdata.csv', 'Data/JC-202210-citibike-tripdata.csv', 'Data/JC-202211-citibike-tripdata.csv', 'Data/JC-202212-citibike-tripdata.csv']

# Create an empty list to store each dataframe
citiBike2022_list = []

# Loop through each file in the list
for file in files_2022:
    # Read the file into a dataframe
    citiBike2022 = pd.read_csv(file)
    
    # Append the dataframe to the list
    citiBike2022_list.append(citiBike2022)
    
# Concatenate all dataframes in the list into one dataframe
citiBike2022 = pd.concat(citiBike2022_list)

# Show the combined dataframe
citiBike2022
```

```python
# List of file names to be combined
files_2023 = ['Data/JC-202301-citibike-tripdata.csv', 'Data/JC-202302-citibike-tripdata.csv', 'Data/JC-202303-citibike-tripdata.csv', 'Data/JC-202304-citibike-tripdata.csv', 'Data/JC-202305-citibike-tripdata.csv', 'Data/JC-202306-citibike-tripdata.csv', 'Data/JC-202307-citibike-tripdata.csv']
# Create an empty list to store each dataframe
citiBike2023_list = []

# Loop through each file in the list
for file in files_2023:
    # Read the file into a dataframe
    citiBike2023 = pd.read_csv(file)
    
    # Append the dataframe to the list
    citiBike2023_list.append(citiBike2023)
    
# Concatenate all dataframes in the list into one dataframe
citiBike2023 = pd.concat(citiBike2023_list)

# Show the combined dataframe
citiBike2023
```

```python
# Dataframes into one
CitiBike2022_2023 = pd.concat([citiBike2022, citiBike2023], axis=0)

#Display the results
CitiBike2022_2023
```

```python
# Save the dataframe to a CSV file
CitiBike2022_2023.to_csv('data/citibike.csv', index=False)
```
