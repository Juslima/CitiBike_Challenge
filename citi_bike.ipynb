```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined
files_2023 = ['Data/JC-202301-citibike-tripdata.csv', 'Data/JC-202302-citibike-tripdata.csv', 'Data/JC-202303-citibike-tripdata.csv', 'Data/JC-202304-citibike-tripdata.csv', 'Data/JC-202305-citibike-tripdata.csv', 'Data/JC-202306-citibike-tripdata.csv', 'Data/JC-202307-citibike-tripdata.csv']
# Create an empty list to store each dataframe
citiBike2023_list = []

# Loop through each file in the list
for file in files_2023:
    # Read the file into a dataframe
    citiBike2023 = pd.read_csv(file)
    
    # Append the dataframe to the list
    citiBike2023_list.append(citiBike2023)
    
# Concatenate all dataframes in the list into one dataframe
citiBike2023 = pd.concat(citiBike2023_list)

# Show the combined dataframe
citiBike2023
```

```python
# Save the dataframe to a CSV file
citiBike2023.to_csv('CitiBike.csv', index=False)
```
